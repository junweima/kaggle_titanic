```python
import pandas as pd
import numpy as np
from numpy import nan
```

```python
train = pd.read_csv('./data./train.csv')
test = pd.read_csv('./data/test.csv')
```

```python
train.head(3)
```

```python
test.head(3)
```

```python
full_data = [train, test]
```

```python
for dataset in full_data:
    dataset['Name Length'] = dataset['Name'].apply(len)
    dataset['Has Cabin'] = dataset['Cabin'].apply(lambda x: 0 if type(x)==float else 1)
    dataset['Family Size'] = dataset['SibSp'] + dataset['Parch'] + 1
    
    # https://stackoverflow.com/questions/44890713/selection-with-loc-in-python
    dataset['isAlone'] = 0
    dataset.loc[dataset['Family Size']==1, 'isAlone'] = 1
    
    dataset['CategoricalFare'] = pd.qcut(dataset['Fare'], 4)
    
    age_avg = dataset['Age'].mean()
    age_std = dataset['Age'].std()
    age_null_count = dataset['Age'].isnull().sum()
    age_null_random_list = np.random.randint(age_avg - age_std, age_avg + age_std, size=age_null_count)
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list
    dataset['Age'] = dataset['Age'].astype(int)
    dataset['CategoricalAge'] = pd.cut(train['Age'], 5)
    
```
